# Task 2: What times of the day can we bring more till operators to the tills ?

```python
from matplotlib import pyplot as plt

# Convert 'Date_Time' column to datetime format
df['Date_Time'] = pd.to_datetime(df['Date_Time'])

# Extract the hour from the 'Date_Time' column
df['Hour'] = df['Date_Time'].dt.hour

# Group by hour and count the number of transactions
transactions_by_hour = df.groupby('Hour')['Transaction No'].count()

# Print the number of transactions by hour
print("Transactions by Hour:")
print(transactions_by_hour)

# Plotting the transactions by hour
plt.figure(figsize=(10, 6))
transactions_by_hour.plot(kind='bar', color='skyblue')
plt.title("Number of Transactions by Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Convert the dataframe to a list of lists format
transactions = df.groupby('Transaction No')['Product Name'].apply(list).tolist()

# Encode the transaction data
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
trans_df = pd.DataFrame(te_ary, columns=te.columns_)

# Compute frequent itemsets using Apriori algorithm with a support of 0.3
frequent_itemsets = apriori(trans_df, min_support=0.3, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Sort the rules by confidence in descending order
sorted_rules = rules.sort_values(by='confidence', ascending=False)

# Print the top rules
sorted_rules
```
